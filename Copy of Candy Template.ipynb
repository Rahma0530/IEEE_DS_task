### 4- drop all the useless columns and  rows with too much NaN ,
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN
- rename columns for easy access

```python
df=df.drop(['Q7: JOY OTHER','Q8: DESPAIR OTHER','Click Coordinates (x, y)','Unnamed: 113','Internal ID','Q9: OTHER COMMENTS'],axis=1)
```

```python
df = df.rename(columns = {'Q1: GOING OUT?' : 'going out', 'Q2: GENDER' : 'gender', 'Q3: AGE': 'age', 'Q4: COUNTRY' : 'country','Q5: STATE, PROVINCE, COUNTY, ETC':'state','Q10: DRESS' : 'dress', 'Q11: DAY': 'day','Q12: MEDIA [Daily Dish]':'Daily Dish','Q12: MEDIA [Science]':'Science','Q12: MEDIA [ESPN]':'ESPN','Q12: MEDIA [Yahoo]':'Yahoo'})
```

### 5 - clean `going out` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df['going out'].value_counts()
```

```python
df['going out'].isna().sum()
```

```python
df['going out'].fillna('Yes',inplace = True)
print('Number of nan values is : ',df['going out'].isna().sum())
print(df['going out'].value_counts())
```

### 6 - clean ` gender` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df['gender'].value_counts()
```

```python
df['gender'].isna().sum()
```

```python
df['gender'] = df['gender'].replace("Other","I'd rather not say")
```

```python
df['gender'].fillna(df['gender'].mode()[0],inplace = True)
```

```python
print('Number of nan values is : ',df['gender'].isna().sum())
print(df['gender'].value_counts())
```

### 7 - clean ` age` column
> tip : note the type of age column , and NaN values
 - try to use `to_numeric`
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
df['age']=pd.to_numeric(df['age'],errors='coerce')
```

```python
print(df['age'].unique())
```

```python
df['age'].describe()
```

```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
iqr = q3 - q1
upper_bound = q3 + 1.5 * iqr
lower_bound = q1 - 1.5 * iqr
print(upper_bound,lower_bound)
```

```python
for i in df['age']:
    if(i>74 or i<10):
        df['age']=df['age'].replace(i,np.nan)
```

```python
print(df['age'].unique())
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['age']=df['age'].astype(int)
```

```python
print('Number of nan values is : ',df['age'].isna().sum())
print(df['age'].value_counts())
```
